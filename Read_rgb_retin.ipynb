```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
import cv2

# فرض می‌کنیم تصویر I قبلاً خوانده شده است
# مثال: I = Image.open("path_to_your_image.tif")

# تبدیل تصویر به آرایه NumPy
I_array = np.array(I)

# بررسی اینکه تصویر رنگی است یا خاکستری
if len(I_array.shape) == 3:  # تصویر رنگی
    # استخراج کانال سبز
    green_channel = I_array[:, :, 1]  # کانال دوم
else:  # تصویر خاکستری
    green_channel = I_array  # خود تصویر به عنوان ورودی استفاده می‌شود

# اعمال CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
enhanced_img = clahe.apply(green_channel)

# تبدیل به نوع داده double
Img = enhanced_img.astype(np.float64)

# نمایش تصویر
plt.imshow(Img, cmap='gray')
plt.axis('off')
plt.title("Processed Image")
plt.show()
```
